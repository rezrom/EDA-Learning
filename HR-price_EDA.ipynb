## DATA Exploration

```python
plt.figure(figsize = (12, 8))

plt.title('Attrition ?', size = 15, fontweight = 'bold')

labels = 'Work', 'No Work'

plt.pie(df1['Attrition'].value_counts(), autopct = '%.0f%%', 
        textprops = {'fontsize' : 17, 'color' : 'white'}, explode = (0.1,0))

plt.legend(labels, loc='lower left', fontsize = 13)

plt.show()
```

# The Piechart shows high attrition percentage of the employee which >10% attrition rate according to https://insightglobal.com/blog/employee-attrition-rate-how-to-calculate-improve/

```python
plt.figure(figsize = (12,6))
sns.heatmap(df.select_dtypes(['int64', 'object', 'float64']).corr('pearson'), annot = True)
```

```python
df_con=df1.select_dtypes(["int64"])
for col in df_con :

    plt.figure(figsize = (10,5))
    
    sns.distplot(df_con[col][df1["Attrition"]=='No'], bins=[np.percentile(df_con.loc[:,col],0),np.percentile(df_con.loc[:,col],25), np.percentile(df_con.loc[:,col],50),np.percentile(df_con.loc[:,col],75),np.percentile(df_con.loc[:,col],100)], color ='g', label = "no attrition")
    sns.distplot(df[col][df1["Attrition"]=='Yes'], bins=[np.percentile(df_con.loc[:,col],0),np.percentile(df_con.loc[:,col],25), np.percentile(df_con.loc[:,col],50),np.percentile(df_con.loc[:,col],75),np.percentile(df_con.loc[:,col],100)], color ='r', label = "attrition" )
    plt.xlabel(col, fontsize = 12)
    plt.title('Distribution of ' + col, fontsize = 22)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

### age, joblecel, jobsatisfactaion, worklifebalance showed decrease in attrition rate along with higher data
### employees with higher joblevel, income, work involvement, job satisfacation

```python
for col in ('EnvironmentSatisfaction', 'JobInvolvement', 'JobSatisfaction', 'RelationshipSatisfaction', 'PerformanceRating', 'WorkLifeBalance'):
    plt.figure(figsize = (8,4))
    ax  = sns.countplot(y= col, data = df, hue='Attrition')
    
    total = len(df[col])
    for val in ax.patches:
        pct = '{:.2f}%'.format(100 * val.get_width()/total)
        xpos = val.get_x() + val.get_width() +1.5
        ypos = val.get_y() + val.get_height() /2 
        ax.annotate(pct, (xpos, ypos), fontsize = 13)
        
    bad = df[col].value_counts()[1]
    medium  = df[col].value_counts()[2]
    high = df[col].value_counts()[3]
    very_high = df[col].value_counts()[4]
    
    plt.ylabel(col, fontsize = 17)
    plt.tight_layout()
    plt.show()
```
